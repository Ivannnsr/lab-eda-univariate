```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = r"C:\Users\ivann\Documents\Data & Analytics\clases\Unit 11 - Statistics\lab-eda-univariate\amz_uk_price_prediction_dataset.csv"

# Cargar el CSV
df = pd.read_csv(file_path)

# Mostrar las primeras filas
print(df.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import skew, kurtosis

# Generar la tabla de frecuencia de categorías
category_counts = df['category'].value_counts()

# Seleccionar un subconjunto de las principales categorías (Top 10)
top_categories_subset = category_counts.head(10)
print("Top categorías más listadas (subset):")
print(top_categories_subset)

# Visualización: Distribución de productos por categoría (gráfico de barras)
plt.figure(figsize=(12, 6))
top_categories_subset.plot(kind='bar', color='skyblue')
plt.xlabel("Categoría")
plt.ylabel("Frecuencia")
plt.title("Distribución de productos por categoría (Subset)")
plt.xticks(rotation=45, ha='right')
plt.show()

# Visualización: Proporción de las categorías más listadas (gráfico de pastel)
plt.figure(figsize=(8, 8))
top_categories_subset.plot(kind='pie', autopct='%1.1f%%', colors=["lightblue", "lightcoral", "lightgreen", "gold", "orchid", "lightpink", "lightgray", "cyan", "orange", "purple"], startangle=140)
plt.title("Proporción de las principales categorías (Subset)")
plt.ylabel("")  # Ocultar la etiqueta del eje Y
plt.show()

# Análisis de precios
mean_price = df['price'].mean()
median_price = df['price'].median()
mode_price = df['price'].mode()[0]

print(f"Precio medio: {mean_price:.2f} GBP")
print(f"Precio mediano: {median_price:.2f} GBP")
print(f"Precio más común (moda): {mode_price:.2f} GBP")

# Medidas de dispersión
variance = np.var(df['price'], ddof=1)
std_dev = np.std(df['price'], ddof=1)
price_range = df['price'].max() - df['price'].min()
iqr = np.percentile(df['price'], 75) - np.percentile(df['price'], 25)

print(f"Varianza: {variance:.2f}")
print(f"Desviación estándar: {std_dev:.2f}")
print(f"Rango de precios: {price_range:.2f} GBP")
print(f"Rango intercuartil (IQR): {iqr:.2f} GBP")

# Visualización: Histograma de precios
plt.figure(figsize=(12, 6))
plt.hist(df['price'], bins=50, color='skyblue', edgecolor='black')
plt.xlabel("Precio (GBP)")
plt.ylabel("Frecuencia")
plt.title("Distribución de precios de productos")
plt.show()

# Visualización: Boxplot de precios
plt.figure(figsize=(8, 6))
plt.boxplot(df['price'], vert=False, patch_artist=True, boxprops=dict(facecolor="lightblue"))
plt.xlabel("Precio (GBP)")
plt.title("Boxplot de distribución de precios")
plt.show()

# Análisis de calificaciones
mean_rating = df['stars'].mean()
median_rating = df['stars'].median()
mode_rating = df['stars'].mode()[0]

print(f"Calificación media: {mean_rating:.2f}")
print(f"Calificación mediana: {median_rating:.2f}")
print(f"Calificación más común (moda): {mode_rating:.2f}")

# Medidas de dispersión para calificaciones
variance_rating = np.var(df['stars'], ddof=1)
std_dev_rating = np.std(df['stars'], ddof=1)
iqr_rating = np.percentile(df['stars'], 75) - np.percentile(df['stars'], 25)

print(f"Varianza de calificación: {variance_rating:.2f}")
print(f"Desviación estándar de calificación: {std_dev_rating:.2f}")
print(f"Rango intercuartil (IQR) de calificación: {iqr_rating:.2f}")

# Forma de la distribución
skewness = skew(df['stars'].dropna())
kurt = kurtosis(df['stars'].dropna())

print(f"Asimetría (Skewness): {skewness:.2f}")
print(f"Curtosis: {kurt:.2f}")

# Visualización: Histograma de calificaciones
plt.figure(figsize=(12, 6))
plt.hist(df['stars'].dropna(), bins=10, color='lightblue', edgecolor='black')
plt.xlabel("Calificación (estrellas)")
plt.ylabel("Frecuencia")
plt.title("Distribución de calificaciones de productos")
plt.show()
```
